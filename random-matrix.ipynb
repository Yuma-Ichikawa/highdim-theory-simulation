```python
import numpy as np
import matplotlib.pyplot as plt

# ランダム行列の生成
n, d = 4000, 2000  # サンプル数 n と特徴量数 d
X = np.random.randn(n, d)  # 平均0、分散1のガウス分布に従うランダム行列

# 分散共分散行列の計算
Sigma = (1 / n) * X.T @ X  # 分散共分散行列

# 固有値を計算
eigenvalues = np.linalg.eigvalsh(Sigma)  # 固有値を計算（昇順）

# プロットの準備
fig, ax = plt.subplots(figsize=(8, 6))  # axオブジェクトを作成

# 固有値分布のヒストグラムをプロット
ax.hist(
    eigenvalues, bins=50, density=True, alpha=0.7, color="blue", label="Empirical"
)

# ラベルやタイトルの設定
ax.set_xlabel("Eigenvalue", fontsize=14)
ax.set_ylabel("Density", fontsize=14)
ax.set_title("Eigenvalue Distribution of Random Matrix", fontsize=16)
ax.legend(fontsize=12)

# グリッドを追加
ax.grid(alpha=0.5)

# プロットの表示
plt.show()
```
